```python
# Write a NumPy function to select values satisfying a specific condition like Values greater than 4 

def values_greater_than_any_given_value(a, any_value):
    x = a > any_value
    satisfying_values = a[x]
    return satisfying_values
Required_values = values_greater_than_any_given_value(a,5)
print(Required_values)    

# Write a NumPy function to select values satisfying mulitple conditions like Values greater than 4 and divisible by 2

def values_satisfying_given_conditions(a):
    x = a > 3
    y = a % 2 == 0
    z = np.logical_and(x,y)

    satisfying_values = a[z]
    return satisfying_values

Required_values = values_satisfying_given_conditions(a)   
print(Required_values)
```
